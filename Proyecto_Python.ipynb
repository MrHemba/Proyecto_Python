Limpieza de datos

```python
cat_cols = [
    "job", "marital", "education", "default", "housing", "loan",
    "contact", "month", "poutcome", "y"
]
def marcar_na(valor):
    if valor == "unknown":
        return pd.NA
    return valor

for c in cat_cols:
    df[c] = df[c].apply(marcar_na)

df[cat_cols] = df[cat_cols].fillna("desconocido")

df["age"] = df["age"].fillna(df["age"].median())


```

```python
#Por recomendacion de readme eliminamos las columnas recomendadas
df = df.drop(columns=["Unnamed: 0", "duration"])
```

Analisis de distribuciones

```python
plt.figure(figsize=(8,4))
sns.histplot(df["age"], bins=20, color="lightblue", edgecolor="black")
plt.title("Distribución de edades de los clientes")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
```

Analisis de campaña

```python
conteo_y = df["y"].value_counts().reset_index()
conteo_y.columns = ["resultado", "cantidad"]

plt.figure(figsize=(6,4))
sns.barplot(data=conteo_y, x="resultado", y="cantidad", color="salmon", edgecolor="black")
plt.title("Distribución de resultados de campaña")
plt.xlabel("Resultado (suscripción)")
plt.ylabel("Cantidad de clientes")
```

```python
#La mayoría de contactos no resultaron en una suscripción
```

```python
def tasa_conversion(grupo):
    total = len(grupo)
    exitosos = len(grupo[grupo["y"] == "yes"])
    prob = exitosos / total
    return prob

#esta funcion de conversion se usara de ahora en adelante 
```

Anaisis de conversion

```python
trab = df.groupby("job").apply(tasa_conversion).reset_index(name="tasa")
trab = trab[trab["job"] != "desconocido"].reset_index(drop=True)

plt.figure(figsize=(9,5))
sns.barplot(data=trab, x="job", y="tasa", color="skyblue", edgecolor="black")
plt.title("Tasa de aceptación por profesión")
plt.ylabel("Porcentaje de aceptación")
plt.xlabel("Profesión")
plt.xticks(rotation=45)
```

```python
trab["tasa"].sum()
```

```python
#Los segmentos con mayor predisposición a la conversión son los Estudiantes y los Jubilados.
```

```python
edu = df.groupby("education").apply(tasa_conversion).reset_index(name="tasa")
edu = edu[edu["education"] != "desconocido"].reset_index(drop=True)

plt.figure(figsize=(8,5))
sns.barplot(data=edu, x="education", y="tasa", color="lightgreen", edgecolor="black")
plt.title("Tasa de aceptación por nivel educativo")
plt.ylabel("Porcentaje de aceptación")
plt.xlabel("Nivel educativo")
plt.xticks(rotation=45)
```

```python
bins = [15,25,35,45,55,65,120]
etq = ["15-24","25-34","35-44","45-54","55-64","65+"]
df["rango_edad"] = pd.cut(df["age"], bins=bins, labels=etq, right=False)

edad = df.groupby("rango_edad").apply(tasa_conversion).reset_index(name="tasa")

plt.figure(figsize=(7,5))
sns.barplot(data=edad, x="rango_edad", y="tasa", color="salmon", edgecolor="black")
plt.title("Tasa de aceptación por grupo de edad")
plt.ylabel("Porcentaje de aceptación")
plt.xlabel("Grupo de edad")
```

```python
#La mayor tasa de éxito se concentra en los extremos de edad: los Jóvenes y Jubilados son los grupos más receptivos.
```

```python
contacto = df.groupby("contact").apply(tasa_conversion).reset_index(name="tasa")
contacto = contacto[contacto["contact"] != "desconocido"].reset_index(drop=True)

plt.figure(figsize=(6,4))
sns.barplot(data=contacto, x="contact", y="tasa", color="orange", edgecolor="black")
plt.title("Tasa de aceptación por canal de contacto")
plt.ylabel("Tasa de aceptación")
plt.xlabel("Canal de contacto")
```

```python
#El canal celular muestra la mayor efectividad en las campañas.
```

```python
pout = df.groupby("poutcome").apply(tasa_conversion).reset_index(name="tasa")
pout = pout[pout["poutcome"] != "desconocido"].reset_index(drop=True)

plt.figure(figsize=(6,4))
sns.barplot(data=pout, x="poutcome", y="tasa", color="violet", edgecolor="black")
plt.title("Tasa de aceptación según resultado previo")
plt.ylabel("Tasa de aceptación")
plt.xlabel("Resultado previo")
```

```python
#Los clientes con resultados previos exitosos tienen más probabilidad de conversión.
```

```python
conv = df[df["y"] == "yes"]["campaign"].value_counts().sort_index().reset_index()
conv.columns = ["llamadas", "conversiones"]

plt.figure(figsize=(8,5))
sns.barplot(data=conv, x="llamadas", y="conversiones", color="lightcoral", edgecolor="black")
plt.title("Número de llamadas necesarias para lograr la aceptación")
plt.ylabel("Aceptación por llamada")
plt.xlabel("Número de llamadas")
```

```python
#La mayoría de conversiones se logra con tres llamadas o menos.
```

Los clientes entre 25 y 45 años, con educación técnica o universitaria,
   y empleos administrativos o de gestión, son los más propensos a contratar depósitos a plazo.

Los contactos por celular resultan más efectivos que los fijos.

La mayoría de conversiones ocurre en tres llamadas o menos.

Los resultados previos positivos aumentan la probabilidad de éxito.

El perfil ideal seria adulto joven, educado, con experiencia positiva previa y contacto por celular.

¿Qué tipo de cliente suscribe el depósito a
plazo, y cuántas llamadas son necesarias para lograrlo? 

El análisis muestra que los clientes con mayor probabilidad de aceptar el depósito a plazo son principalmente jóvenes (15–24 años) y adultos mayores (65+), especialmente aquellos que son estudiantes, jubilados o que cuentan con educación universitaria o cursos profesionales. Además, los clientes que ya tuvieron un resultado exitoso previo en una campaña presentan la tasa de aceptación más alta. El contacto por celular es claramente más efectivo que el teléfono fijo, y la mayoría de conversiones se logra con 1 a 3 llamadas, por lo que insistir más allá de ese punto no genera beneficios.
